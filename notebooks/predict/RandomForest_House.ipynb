```python
import pandas as pd
import numpy as np
from sklearn.impute import KNNImputer
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import GridSearchCV

# Load data
df = pd.read_csv('../../data/batdongsan/numerized/NhaO_numerized-dataset.csv')

# Kiểm tra các giá trị bị thiếu
print(df.isnull().sum())

# Khởi tạo KNN Imputer
imputer = KNNImputer(n_neighbors=5)

# Chọn các cột cần điền giá trị thiếu

columns_to_impute = df.select_dtypes(include=[np.number]).columns
columns_to_round = ['SoToilet',"SoTang", 'SoPhongNgu', 'NoiThat', 'HuongNha', 'HuongBanCong', 'PhapLy']

# Điền giá trị thiếu
df[columns_to_impute] = imputer.fit_transform(df[columns_to_impute])
df[columns_to_round] = df[columns_to_round].round()
# Điền giá trị thiếu

# Kiểm tra lại các giá trị bị thiếu
print(df.isnull().sum())
```

```python
df = df.dropna()
print(df)
```

```python
# df = df.dropna(subset=['MucGia'])
df["USD"] = (df["MucGia"]/25400).round(2)
# df=df.drop(columns="MucGia")
```

```python
df=df.drop(columns=["Unnamed: 0","DiaChi","MucGia","Lat","Long"])
```

```python
# Assuming df is your DataFrame
X = df.iloc[:,0:13].values#features
y = df.iloc[:,13].values# Target variable# Target variable
```

```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import LabelEncoder

# Check for and handle categorical variables
label_encoder = LabelEncoder()
x_categorical = df.select_dtypes(include=['object']).apply(label_encoder.fit_transform)
x_numerical = df.select_dtypes(exclude=['object']).values
x = pd.concat([pd.DataFrame(x_numerical), x_categorical], axis=1).values
# Fitting Random Forest Regression to the dataset
regressor = RandomForestRegressor(n_estimators=10, random_state=0, oob_score=True)

# Fit the regressor with x and y data
regressor.fit(x, y)
```

```python
# Evaluating the model
from sklearn.metrics import mean_squared_error, r2_score

# Access the OOB Score
oob_score = regressor.oob_score_
print(f'Out-of-Bag Score: {oob_score}')

# Making predictions on the same data or new data
predictions = regressor.predict(x)

# Evaluating the model
mse = mean_squared_error(y, predictions)
print(f'Mean Squared Error: {mse}')
r2 = r2_score(y, predictions)
print(f'R-squared: {r2}')
```

```python
from sklearn.preprocessing import StandardScaler

# Standardize the features
scaler = StandardScaler()
x = scaler.fit_transform(x)

# Fit the regressor with standardized data
regressor.fit(x, y)
predictions = regressor.predict(x)
mse = mean_squared_error(y, predictions)
r2 = r2_score(y, predictions)
oob_score = regressor.oob_score_
print(f'Out-of-Bag Score: {oob_score}')
print(f'Mean Squared Error: {mse}')
print(f'R-squared: {r2}')
```
